```python
import csv

results_dict = {}
with open(TN93_FILE, 'r') as data:
    for line in csv.DictReader(data):
        #print(line)
        
        
        # {'NM_022685_2_Rattus_norvegicus_RRAD_and_GEM_like_GTPase_2_Rem2_mRNA_1': {'NM_080726_3_Mus_musculus_rad_and_gem_related_GTP_binding_protein_2_Rem2_mRNA_1': 0.02848115779748788,
        
        if line["ID1"] in results_dict.keys():
            results_dict[line["ID1"]].update({line["ID2"]: float(line["Distance"])})
        else:
            results_dict[line["ID1"]] = {line["ID2"]: float(line["Distance"])}
        #end if
```

```python
results_dict
```

```python

df = pd.DataFrame.from_dict(results_dict)
df
```
